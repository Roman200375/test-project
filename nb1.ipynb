```python
def draw_pyramid(height):
    for i in range(1, height + 1):
        spaces = " " * (height - i)
        stars = "*" * (2 * i - 1)
        print(spaces + stars)

try:
    height = int(input("Введіть висоту піраміди: "))
    draw_pyramid(height)
except ValueError:
    print("Будь ласка, введіть коректне число.")
```

робота з циклами
while - умовний цикл - ітерації буду виконуватись поки умова є істиною (може бути безмежним циклом)

for - із задананою кількістю ітерації (повторів)

```python
from random import randint
n = ["Владислав","Сергій","Василь"]
h = ["Ігри","фотошоп","програмування"]
j = 0
while j < 3:
    print(">>>{name} любить {hobbi}\n".format(name=n[j],hobbi=h[randint(0,2)]))
    j+=1
print("Це вже не входить у цикл, цикл закінчився.\nЦикл While не дуже підходить для визначеної кількості ітерацій.")

# Навіть для for можна йти по різному
for j in range(len(n)):
    print(">>>{name} любить {hobbi}\n".format(name=n[j],hobbi=h[randint(0,2)]))
print("Це вже виконується поза циклом")

print(f"Демонструємо що буде виконувати функція range: {list(range(3))}")

# варіант 2 для for (це вж павильне використаня ащк)
for j in n:
    print(">>> {name} любить {hobbi}\n".format(name=j,hobbi=h[randint(0,2)]))
# Спростимо код ще більше
for j in n:
    print(f">>> {j} любить {h[randint(0,2)]}")
```

пробуємо використати цикли при симуляції кидання кубика

```python
# числа можуть бути тільки від 1 до 5, і ми їх можемо вводити з клавіатури
required_number = 5 #int(input("Введіть бажане значення числа на кубику (від 1 до 6):"))

dice_number = int()
print("Значення яка повинно випасти на кубику відповідає умові від 1 до 6?")
if 1 <= required_number <= 6:
    print("Введено коректні дані, починаємо кидати кубик:")
    while required_number != dice_number: # задаємо умову неістинності
        dice_number = randint(1,6)
        print(f"Кидаємо кубик, випало значення {dice_number}. Отримане значення по відношенню до бажаного буде: {required_number == dice_number}.")
else:
    print(f"Введено не коректне значення: {required_number}")
```

```python
attempts = 4 # Кількість ітерацій
required_number = 6 # число яке має випасти на кубику
dice_number = int()
# якщо ми не будемо використовувати нашу ітераційну змінну у циклі, то можна її назвати _ (просто нижнє підкреслення)
for _ in range(attempts):
    dice_number = randint(1,6)
    print(f"Кидаємо кубик, випало значення {dice_number}. Отримане значення по відношенню до бажаного буде: {required_number == dice_number}.")
    if required_number == dice_number:
        print(f"Умова виконана на {_ + 1} ітерації, виходимо з циклу.")
        break
else:
    print("Задане значення не було отримане.")
```

знайомство з обробниками винятків

```python
# код призведе до помилки бо ми п хочемо викликати метод притаманний лише для int змінних інших класів
ll = [int(2), int(10000), str("10"), float(5000), bytes(1)]
for a in ll:
    print(f"Змінна а має тип{type(a)}.")
    # використання блоку IF для обробки помилок є неправильним!
    if isinstance(a, int): # type(a) == int Це теж саме
        print(f"Кількість байт у зміній а={a.bit_count()}. Вивід через IF")
    else:
        print("Змінна а не є типу int!")
    # для обробки помилок ми повинні використовувати конструкцію try
    try:
        print(f"Кількість байт у змінній а={a.bit_count()}. Вивід через TRY.")
    except AttributeError as e:
        print(f"Виникла помилка, деталі {e}")
    try:
        print(f"Кількість байт у змінній а={int(a).bit_count()}. Вивід через TRY.")
    except ValueError as e:
        print(f"Виникла помилка, деталі {e}\n Тут неможливо порахувати кількісмть байт у змінній.")
```

випадково генеруємо числа де може попадатись число 0, на яке не можна ділити

за допомогою конструкції try ми можемо опрацювати помилку ділення на 0 та просто задати іншу логіку виконання даної операції

```python
li = [randint(-5,5) for _ in range(30)]
print(li)
for i in li:
    try:
        d = 5/i

    except ZeroDivisionError:
        print(f"Було згенеровано число {i} на яке ділити не можна.")
        d = None
    finally:
        print(f"Будемо ділити 5 на число {i}, результат буде: {d}")
```

існують іменні та анонімні/безіменні функції, і лямбда є безіменною!

```python
# після def буде назва функції за якоюми можемо її викликати
def name_of_my_fun(arg:str):
    # Тіло функції
    print(f"У функцію було передано значення: {arg}")
    return True if arg != None else False
    # Після цієї команди більше ніякі операції виконуватись не будуть
    print("Це ніколи не виконається")
name_of_my_fun(1)
print(f"Виклика функції з аргументом: {name_of_my_fun(1)}")
# функція без аргументів
def no_args_fun():
    return True
print(f"Друга функція повернула значення: {no_args_fun()}")
```

```python
# Щодо виклику функції, він здійснюється за допомогою ()
a = no_args_fun

print(a)

def function_call_function(f):
    print("У цю функцію ми передали інакшу функцію  і тепер викликаємо її тут:")
    f()
    print("Передана функція буде викликаною.")

def internal_no_args():
    print("Ця функція буде переданою кудись.")

function_call_function(internal_no_args)
```

а тепер спробуємо зробити Лямбду

```python
# прототипом Лямбди буде така функція
def lambda_prototype(a:int, b:float):
    return  f"Множимо два отримані числа: {a * b}"

a = lambda_prototype # присвоємо НЕ викликану функцією у змінну

print(f"Проста функція {a}, далі йде її виклик:", a(2, 2.5))

# перепишемо цю функцію  через Лямбду
lb = lambda a, b: f"Множимо два отримані числа: {a * b}"

print(f"Ми створили Лямбду {lb}, далі йде її виклик:", lb(2,2.5))
```

```python
# приклад використання Лямбд
la = lambda i: [i+j if i > 10 else i for j in range(i)]
la(11)
```
